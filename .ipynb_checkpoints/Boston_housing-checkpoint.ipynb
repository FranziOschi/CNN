```python
from keras.datasets import boston_housing

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
# load data
(X_train, Y_train), (X_test, Y_test) = boston_housing.load_data()
```

```python
plt.hist(Y_train)
plt.xlabel('House prices [k$]')
```

```python
# log transformation of target variable
Y_train = np.log1p(Y_train)
Y_test = np.log1p(Y_test)

plt.hist(Y_train)
plt.xlabel('House prices [k$]')
plt.title('Log transformation')
```

#### Analysis of features

```python
# univariate analysis with pearson correlation coefficient in title
univar_pcc = np.zeros(X_train.shape[1])

plt.figure(figsize = (10,10))
plt.subplots_adjust(wspace = 0.5, hspace = 0.5)
for i in range(X_train.shape[1]):
    plt.subplot(4,4,i+1)
    plt.scatter(Y_train, X_train[:,i])
    pcc = stats.pearsonr(Y_train, X_train[:,i])
    plt.title(np.round(pcc[0],2))
    
    univar_pcc[i] = pcc[0]
```

```python
def extract_X_tresh(tresh_pcc, univar_pcc):
    abs_univar_pcc = abs(univar_pcc)
    ix_tresh_pcc = np.where(abs_univar_pcc > tresh_pcc)[0]
    
    X_train_tresh = np.vstack([X_train[:,ix] for ix in ix_tresh_pcc]).T
    X_test_tresh = np.vstack([X_test[:,ix] for ix in ix_tresh_pcc]).T
    
    return X_train_tresh, X_test_tresh
```

```python
def lin_regr(X_train, Y_train, X_test, Y_test):
    lin_model = LinearRegression()
    lin_model.fit(X_train, Y_train)
    
    y_train_predict = lin_model.predict(X_train)
    rmse_train = (np.sqrt(mean_squared_error(np.exp(Y_train), np.exp(y_train_predict))))
    #rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
    r2_train = r2_score(Y_train, y_train_predict)
    
    # model evaluation for testing set
    y_test_predict = lin_model.predict(X_test)
    rmse_test = (np.sqrt(mean_squared_error(np.exp(Y_test), np.exp(y_test_predict))))
    #rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
    r2_test = r2_score(Y_test, y_test_predict)
    
    return rmse_train, r2_train, rmse_test, r2_test
```

```python
# introduce treshold value for pcc

tresh_pcc = np.arange(0., 0.8, 0.1)

rmse_train_ar = np.zeros(tresh_pcc.shape)
rmse_test_ar = np.zeros(tresh_pcc.shape)
r2_train_ar = np.zeros(tresh_pcc.shape)
r2_test_ar = np.zeros(tresh_pcc.shape)

for i, t in enumerate(tresh_pcc):
    X_train_tresh, X_test_tresh = extract_X_tresh(t, univar_pcc)
    
    rmse_train, r2_train, rmse_test, r2_test = lin_regr(X_train_tresh, Y_train, X_test_tresh, Y_test)
    
    rmse_train_ar[i] = rmse_train
    r2_train_ar[i] = r2_train
    rmse_test_ar[i] = rmse_test
    r2_test_ar[i] = r2_test
```

```python
plt.subplot(1,2,1)
plt.plot(tresh_pcc, rmse_train_ar)
plt.plot(tresh_pcc, rmse_test_ar)
plt.subplot(1,2,2)
plt.plot(tresh_pcc, r2_train_ar)
plt.plot(tresh_pcc, r2_test_ar)
```

```python
lin_model = LinearRegression()
lin_model.fit(X_train_tresh, Y_train)
```

```python
# model evaluation for training set
y_train_predict = lin_model.predict(X_train_tresh)
rmse = (np.sqrt(mean_squared_error(np.exp(Y_train), np.exp(y_train_predict))))
#rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
r2 = r2_score(Y_train, y_train_predict)

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")

# model evaluation for testing set
y_test_predict = lin_model.predict(X_test_tresh)
rmse = (np.sqrt(mean_squared_error(np.exp(Y_test), np.exp(y_test_predict))))
#rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```

```python
# biivariate analysis with pearson correlation coefficient in title

bivariate_pcc = np.zeros((X_train.shape[1], X_train.shape[1]))

for i in range(X_train.shape[1]):
    for j in range(X_train.shape[1]):
        pcc = stats.pearsonr(X_train[:,i], X_train[:,j])
        bivariate_pcc[i, j] = pcc[0]
```

```python
plt.figure(figsize = (8,8))
plt.imshow(bivariate_pcc, interpolation='nearest')
plt.colorbar()
```
